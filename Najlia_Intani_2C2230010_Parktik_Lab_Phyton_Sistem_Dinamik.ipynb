 **1.2.1 Tutorial One: A Tutorial Introduction to Sympy**

 Sympy is a computer algebra system and a Python library for symbolic
 mathematics written entirely in Python. The following tutorial has been de
signed to allow new users to become familiar with the commands by means
 of example. For more detailed information please refer to sympy’s document
 pages at:
 http://docs.sympy.org/latest/index.html

 The following command lines should be typed in the console window. There
 is no need to copy the comments, they are there to help you

```python
#BY: NAJLIA CUTE

#1 Approximate decimal.
print(2/3 + 4/5)

#2 To work with fractions.
from fractions import Fraction

#3 Symbolic answer.
print(Fraction(2, 3) + Fraction(4, 5))  # Symbolic answer.

#4 Square root.
print(sqrt(16))

#5 Trigonometric function.
print(sin(pi))

#6 Import everything from sympy library into global scope.
from sympy import *

#7 Declare x and y symbolic.
x, y = symbols('x y')

#8 Factorize.
print(factor(x**3 - y**3))

#9 Expand the last result.
print(expand(x**3 - y**3))

#10 Simplify an expression.
print(factor((x**3 - y**3) / (x - y)))

#11 Partial fractions.
print(apart(1 / ((x + 2) * (x + 1))))

#12 Simplify a trig expression.
print(trigsimp(cos(x) - cos(x)**3))

#13 Limits.
print(limit(1/x, x, oo))

#14 Total differentiation.
print(diff(x**2 - 3*x + 6, x))

#15 Partial differentiation.
print(diff(x**3 * y**5, x, y, 3))

#16 Indefinite integration.
print(integrate(sin(x) * cos(x), x))

#17 Definite integration.
print(integrate(exp(-x**2 - y**2), (x, 0, oo), (y, 0, oo)))

#18 Taylor series expansion.
print((exp(x) * cos(x)).series(x, 0, 10))

#19 An infinite sum.
print(summation(1 / x**2, (x, 1, oo)))

#20 Solving equations roots
print(solve(x**5 - 1, x))

#21 Solving simultaneous equations.
print(solve([x + 5 * y - 2, -3 * x + 6 * y - 15], [x, y]))

#22 Note that 1j=I can be used too.
z1 = 3 + 1 * I
z2 = 5 - 4 * I

#23 Simple complex arithmetic.
print(2 * z1 + z1 * z2)

#24 Complex conjugate.
print(conjugate(z1))

#25 The argument of z1.
print(arg(z1))

#26 The modulus of z1.
print(abs(z1))

#27 The real part.
print(re(z1))

#28 The imaginary part.
print(im(z1))

#29 Express in form x+iy.
print(exp(I * z1).expand(complex=True))

#30 Two 2x2 matrices.
A = Matrix([[1, -1], [2, 3]])
B = Matrix([[0, 2], [3, 3]])

#31 Matrix algebra.
print(2 * A + 3 * A * B)

#32 Acces the first row.
print(A.row(0))

#33 The transpose of a matrix.
print(A.T)

#34 Acces the second column of A.
print(A.T.row(1))

#35 The element in row 1, column 2.
print(A[0, 1])

#36 The inverse of a matrix.
print(A ** (-1))

#37 The power of a matrix.
print(A ** 5)

#38 The determinant of A
print(A.det())

#39 A 3x3 matrix of zeros.
print(zeros(3, 3))

#40 A 1x5 matrix of ones.
print(ones(1, 5))

#41 A 3x3 matrix.
C = Matrix([[2, 1, 0], [-1, 4, 0], [-1, 3, 1]])

#42 The row reduces achelon form.
print(C.rref())

#43 The eigenvalues of C.
print(C.eigenvals())

#44 The eigenvectors of C.
print(C.eigenvects())

#45 Declare s,t,w symbolic.
s, t, w = symbols('s t w')

#46 Laplace transform.
laplace_transform(t**3, t, s)

#47 Inverse transform.
inverse_laplace_transform(6/s**4, s, t)

#48 Fourier transform
fourier_transform(-2 * pi * abs(t), t, w)

#49 Inverse transform.
inverse_fourier_transform(1/(pi * w**2), w, t)

#50 Quits Iphyton console.
quit
```

 **1.2.2 Tutorial Two: A Tutorial Introduction to Numpy
 and Matplotlib**

 Numpy’s main object is the homogeneous multidimensional array and al
lows Python to compute with vectors and matrices. Matplotlib is a Python
 2-dimensional plotting library used to generate bar charts, error charts, his
tograms, plots, power spectra, and scatterplots, for example. When combin
ing the pyplot module with IPython a MATLAB-like interface is provided
 and the user can control axes properties, font properties, and line styles via
 an object oriented interface. For a more detailed reference guide to numpy,
 readers are directed to:

 https://docs.scipy.org/doc/numpy/reference/

 and an introduction to matplotlib is given here:

 https://matplotlib.org

 The following command lines provide a concise introduction to numpy and
 matplotlib by means of example. The following command lines should be
 typed in the IPython console window. There is no need to copy the comments,
 they are there to help you.

```python
#BY: NAJLIA CUTE

#1 Import numpy into the np
# namespace
import numpy as np

#2 A 1d array [0 1 2 3 4]
a=np.array(5)

#3 A 2d array [[0 1 2],[3 4 5]].
b = np.arange(6).reshape(2,3)

#4 A 2d array.
A = np.array([[1, 1], [0, 1]])

#5 A 2d array.
B = np.array([[2, 0], [3, 4]])

#6 Elementwise product [[2 0], [0 4]].
print(A*B)

#7 Matrix product [[5 4],[3 4]].
print(A.dot(B))

#8 Matrix product [[5 4],[3 4]].
print(np.dot(A, B))

#9 A 2d array.
c=np.arange(12).reshape(3, 4)

#10 Sum each column.
print(c.sum(axis=0))

#11 Minimum of each row.
print(c.min(axis=1))

#12 Cumulative sum for each row.
print(c.cumsum(axis=1))

#13 Gives array([ 0.,0.5,1.,1.5,2.]).
print(np.linspace(0, 2, 5))

#14 Simple plots with Python and matplotlib

#15 Import numpy.
from numpy import *

#16 Import pyplot.
import matplotlib.pyplot as plt

#17 Set up the domain.
x = np.linspace(-2, 2, 50)

#18 A function of x.
y = x**2

#19 A basic plot.
plt.plot(x, y)

#20 Two plots on one graph

#21
t = np.linspace(0, 100, 1000) #Set the domain.
p1 = np.exp(-0.1 * t) * np.cos(t) # Two functions.
p2 = np.cos(t)
plt.plot(t, p1)
plt.plot(t, p2) # Plot two curves.
plt.show()

#22 Plot with a legend

#23
plt.plot(t, p1, label='p1')
plt.plot(t, p2, label='p2')
plt.legend()
plt.show()

#24 Plot with labels and title

#25
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x^2')
plt.plot(x, y)
plt.show()

#26 Change line width and colour

#27
plt.plot(x,y,color = 'green', linewidth = 4)
plt.show()

#28 Plotting implicit functions

#29
x,y=np.mgrid[-5:5:100j,-5:5:100j] #A grid of x,y values.
z=x**2/4+y**2 # A function of two variables.
plt.contour(x,y,z,levels=[1]) # The curve x**2/4+y**2=1.
plt.show()

#30 A parametric plot

#31
t=np.linspace(0, np.pi, 100) # A set of t values.
x = 0.7*np.sin(t+1)*np.sin(3*t) # A set of x values.
y = 0.7*np.cos(t+1)*np.sin(3*t) # A set of y values.
plt.plot(x, y) #The 2D parametric plot.
plt.show()

#32 Quits IPython console.
quit
```
